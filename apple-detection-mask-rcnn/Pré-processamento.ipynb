Este notebook possui o único propósito de organizar a estrutura de pasta corretamente. Este notebook pode ser executado da máquina localmente, pois não exige alto poder computacional. Siga os seguintes passos:
- Faça o download da base de dados através deste link: [apple_detection_dataset.zip](https://github.com/OlafenwaMoses/AppleDetection/releases/download/v1/apple_detection_dataset.zip)
- Descompacte o arquivo no mesmo diretório que este notebook se encontra, observe que o nome da pasta é apple_dataset
- Execute todas as células a seguir
- Após a execução de todas as células, faça o upload do diretório data-apple para algum drive de sua preferência.

```python
import os
from os import listdir
```

```python
path = './apple_dataset/train/images/'

desordenados = listdir(path)


ordenados_images = './data-apple/images/'
ordenados_annotations = './data-apple/annotations/'

i = 0
for j in desordenados:
    try:
        if(i < 10):
            new_name_image = 'apple-00' + str(i) + '.jpg'
        elif(i >= 10 and i < 100):
            new_name_image = 'apple-0' + str(i) + '.jpg'
        else:
            new_name_image = 'apple-' + str(i) + '.jpg'
        os.rename(path + j, ordenados_images + new_name_image)
    except (IOError, SyntaxError):
        print('Bad file:', j)
    i+=1
 
```

```python
path = './apple_dataset/train/annotations/'

desordenados = listdir(path)


ordenados_images = './data-apple/images/'
ordenados_annotations = './data-apple/annotations/'

i = 0
for j in desordenados:
    try:
        if(i < 10):
            new_name_xml = 'apple-00' + str(i) + '.xml'
        elif(i >= 10 and i < 100):
            new_name_xml = 'apple-0' + str(i) + '.xml'
        else:
            new_name_xml = 'apple-' + str(i) + '.xml'
        os.rename(path + j, ordenados_annotations + new_name_xml)
    except (IOError, SyntaxError):
        print('Bad file:', j)
    i+=1
```
